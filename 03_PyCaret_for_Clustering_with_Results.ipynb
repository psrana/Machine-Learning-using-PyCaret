# **4. Clustering - Part 2 (Apply Data-Preprocessing)**
---

### **Read the Dataset**

````python
from pycaret.clustering import *
from pycaret.datasets import get_data

jewelleryDataSet = get_data('jewellery')
````

### **4.1 Model Performance using Data Normalization**

````python
setup(data = jewelleryDataSet, normalize = True, normalize_method = 'zscore', silent=True)
x = create_model('kmeans', num_clusters = 4)
````

### **4.2 Model Performance using Transformation**


````python
setup(data = jewelleryDataSet, transformation = True, transformation_method = 'yeo-johnson', silent=True)
x = create_model('kmeans', num_clusters = 4)
````

### **4.3 Model Performance using Transformation + Normalization**

````python
setup(data = jewelleryDataSet, transformation = True, normalize = True, normalize_method = 'zscore', 
      transformation_method = 'yeo-johnson', silent=True)
x = create_model('kmeans', num_clusters = 4)
````

````python
setup(data = jewelleryDataSet, pca = True, pca_method = 'linear', silent=True)
x = create_model('kmeans', num_clusters = 4)
````

### **5.2 Density-Based Spatial Clustering**

````python
from pycaret.datasets import get_data
from pycaret.clustering import *

jewelleryDataSet = get_data('jewellery')
setup(data = jewelleryDataSet, silent=True)
dbscanModel = create_model('dbscan')
````

### **5.2.1 Assign Model - Assign the labels to the dataset**

````python
dbscanModelPrediction = assign_model(dbscanModel)
dbscanModelPrediction

# Noisy samples are given the label -1 i.e. 'Cluster -1'
````

### **Key Points**

- num_clusters not required for some of the clustering Alorithms (Affinity Propagation ('ap'), Mean shift
  clustering ('meanshift'), Density-Based Spatial Clustering ('dbscan') and OPTICS Clustering ('optics')). 
- num_clusters param for these models are automatically determined.
  
- When fit doesn't converge in Affinity Propagation ('ap') model, all datapoints are labelled as -1.
  
- Noisy samples are given the label -1, when using Density-Based Spatial  ('dbscan') or OPTICS Clustering ('optics'). 
  
- OPTICS ('optics') clustering may take longer training times on large datasets.
